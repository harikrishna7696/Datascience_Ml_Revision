```python
iris = sns.load_dataset('iris')
```

```python
iris.plot()
# you can the insights of the data, min and max for flower features
```

```python
# If you want to increase the figure size
iris.plot(figsize=(10, 10))
# if you absorve the below graph is bigger than previous graph
```

```python
# you can also give the small size graph
iris.plot(figsize=(5, 5))
```

```python
# let's give the title of the graph
iris.plot(figsize=(8, 8), title='Iris Features')
```

```python
# you can also give the axis value
iris.plot(figsize=(8, 8), title='Iris Data', xlabel='X(rows)', ylabel='Y(Values)')
```

```python
# for better understand, we can create the graph with grid
iris.plot(figsize=(8, 8), grid=True, title='Iris Data', xlabel='X (rows)', ylabel='Y (Values)')
```

```python
# another way to give the axis values
gf = iris.plot(grid=True)
gf.set_title('Iris Data')
gf.set_xlabel('X')
gf.set_ylabel('y')
```

```python
"""Bar Graph: It allows you to compare different sets of data among different groups easily. 
It instantly demonstrates this relationship using two axes, where the categories are on one axis and the various values are on 
the other. A bar graph can also illustrate important changes in data throughout a period of time"""
```

```python
# we have two differents to create bar one is this, another demonstrance in line
iris.iloc[0][:4].plot(kind='bar')
```

```python
# selecting the first and first 4 columns
# this is another way to create bar graph
iris.iloc[0, :4].plot.bar(color='red') # give the color code or name using the color attribute
```

```python
iris.iloc[:2][:4].plot.bar()
# what below visualization means, taking 2 rows and 4 columsn
# we have four features and different values
# You can find the visualization of two rows, to get idea about values
```

```python
bar_gf = iris.iloc[:4, :4].plot(kind="bar")
bar_gf.set_title('Iris Data')
bar_gf.set_xlabel('X')
bar_gf.set_ylabel('y')
```

```python
# we can also create the horizontal bar graph
iris.iloc[:2][:4].plot(kind='barh')
```

```python
iris.iloc[:2][:4].plot.barh()
```

```python
"""Histogram: The histogram is a popular graphing tool. 
It is used to summarize discrete or continuous data that are measured on an interval scale.
It is often used to illustrate the major features of the distribution of the data in a convenient form"""
```

```python
iris.plot(kind='hist')
# here it taking values as x axis, it covered how much that are min to max
```

```python
iris.plot(kind='hist', stacked=True, bins=100)
# stack it overlap the features, if one is greater than other it will show color, 
# you can observe in the legend.
```

```python
# for horizontal histogram
iris.plot(kind='hist', orientation='horizontal', bins=100, figsize=(10, 10), stacked=True)
# remaing title axis is same as above plots
```

```python
# to give darker or lighter so you can give the alpha value
iris.plot.hist(alpha=0.1)
```

```python
iris.plot.hist(alpha=1.0)
# this alpha is applicable for maximum graphs
```

```python
"""Scatter Plot: A scatter plot is a type of data visualization that shows the relationship between different variables. 
This data is shown by placing various data points between an x- and y-axis. """
```

```python
# In above graphs the x and y taken default, In some graph it taken indexes as x and values are y axis, vice versa.
# but in scatter we need to give x and y
iris.plot.scatter(x='sepal_length', y='sepal_width', color='red')
```

```python
iris.plot.scatter(x='sepal_length', y='sepal_width', color='red', s=50)
# we can increase the size using s=size
```

```python
# there is another graph like scatter plot, which we call hexbin
iris.plot.hexbin(x='sepal_length', y='sepal_width', gridsize=10)
# here instead of s we use gridsize to give the size of plot
# if give higher value, it will disappear, or it will show small
```

```python
"""Pie Chart: Pie charts are often used to represent sample data—with data points belonging 
to a combination of different categories. 
Each of these categories is represented as a “slice of the pie.” """
```

```python
iris.iloc[0][:4].plot.pie()
```
