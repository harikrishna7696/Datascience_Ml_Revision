```python
# Cufflinks: It will create the connection between plotly and Pandas, so you can directly apply visualization on dataframes
```

```python
from plotly.offline import iplot #when we are working in notebook offline, it will show the graph
import plotly as py
import plotly.tools as tls
import cufflinks as clf
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# lets create our first ploty plot sounds good right
iris.iplot()
```

```python
# so we got error because we are doing in notebook which is offline
py.offline.init_notebook_mode(connected=True)
clf.go_offline()
```

```python
iris.iplot()
# Boom! we got the chart 
```

```python
# lets get bigger data for applying more visualizations
titanic =  sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.iplot(x='fare', y='age', mode='markers', legend=True)
```

```python
scatter_py = iris.iplot(kind='scatter', x='sepal_length', y='petal_length', mode='markers', legend=True,
                        title='Relation between sepal_length and petal width', xTitle='sepal_length', yTitle='petal_length')
```

```python
# you can change the plot background color
clf.set_config_file(theme='solar')
iris.iplot(kind='scatter', x='petal_width', y='sepal_width', mode='markers')
```

```python
# bar graph
titanic.iplot(kind='bar', y='survived', x='sex', bins=10)
```

```python
iris.iplot(kind='box')
```

```python
titanic.iplot(kind='box')
```

```python
# scatter 3d
iris.iplot(x='sepal_length', y='petal_length', z='sepal_width', kind='scatter3d', xTitle='sepal_length', yTitle='petal_length', zTitle='sepal_width')
```
